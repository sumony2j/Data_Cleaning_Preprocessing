## Data Cleaning & Preprocessing

### Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler,OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer
```

### Load Dataset

```python
df = pd.read_csv('./Data.csv')
```

```python
df
```

```python
df.corr()
```

```python
df.isnull().sum()
```

```python
Imputer = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0)
```

```python
df.iloc[:,1:3]=Imputer.fit_transform(df.iloc[:,1:3])
```

```python
df
```

```python
Imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
df.iloc[:,1:3]=Imputer.fit_transform(df.iloc[:,1:3])
```

```python
df
```

```python
MinMax=MinMaxScaler()
df.iloc[:,1:3]=MinMax.fit_transform(df.iloc[:,1:3])
```

```python
df
```

```python
St = StandardScaler()
df.iloc[:,1:3]=St.fit_transform(df.iloc[:,1:3])
```

```python
df
```

```python
df.skew()
```

```python
df.kurtosis()
```

```python
Ordinal = OrdinalEncoder()
df.iloc[:,[0,3]]=Ordinal.fit_transform(df.iloc[:,[0,3]])
```

```python
df.index
```

```python
ct = OneHotEncoder(sparse=False)
x=pd.DataFrame(ct.fit_transform(df.iloc[:,[0,3]]))
x.index=df.index
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
df = np.array(ct.fit_transform(df))
print(df)
```
